```python
#Once created the clone of GIT-HUB repository then,
#Required libraries for the program

import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path="C:/Users/AGM MSME PNB/OneDrive/Desktop/Sashidhar/Data scince/Projects/Phonepe Pulse Data Visualization/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list
#print(Agg_state_list)
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}
print(type(clm))
for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transacion_type'].append(Name)
              clm['Transacion_count'].append(count)
              clm['Transacion_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Trans=pd.DataFrame(clm)
Agg_Trans
```

```python
#Once created the clone of GIT-HUB repository then,
#Required libraries for the program
#map based user data show
import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path="C:/Users/AGM MSME PNB/OneDrive/Desktop/Sashidhar/Data scince/Projects/Phonepe Pulse Data Visualization/pulse/data/map/user/hover/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list
#print(Agg_state_list)
#Agg_state_list--> to get the list of states in India
clm_usr={'State':[], 'Year':[],'Quater':[],'district':[], 'registeredUsers':[], 'appOpens':[]}
for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            if D['data']['hoverData'] is not None:
                for z in D['data']['hoverData']:
                    #print(z)
                    district=z
                    s=D['data']['hoverData'][z]
                    #print(s ,s['registeredUsers'])
                    count=s['registeredUsers']
                    appOpens=s['appOpens']
                    clm_usr['district'].append(district)
                    clm_usr['registeredUsers'].append(count)
                    clm_usr['appOpens'].append(appOpens)
                    clm_usr['State'].append(i)
                    clm_usr['Year'].append(j)
                    clm_usr['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_users=pd.DataFrame(clm_usr)
Agg_users
```
